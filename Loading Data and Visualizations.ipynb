Here is an example of the waveform data that we will use. This is one instance of a Ventricular Tachycardia alarm.

In order to work with the data using sklearn, we'll change the amplitude data from the records into a pandas dataframe.

There are four columns that we can observe here: II, aVR, V, and PLETH
1. This is the data we'll use to classify the different instances of Ventricular Tachycarida alarms. II (a.k.a ECG Lead II) is useful for dectecting heart rate, arrythmias, and (in our case) ventrical tachycardias. II is found using the electrical activity between electrodes on the right arm and left leg.
2. aVR (Augmented Vector Right) is another ECG lead that views the heart from the right shoulder perspective.
3. V (Precordial Lead) is a chest ECG lead placed directly on the patient’s chest that provides a different angle on heart activity than limb leads.
4. PLETH (Plethysmography Signal) tracks blood volume changes in the finger using infrared light in a pulse oximeter.
